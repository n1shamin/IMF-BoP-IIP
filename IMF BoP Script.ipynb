```python
country = pd.read_excel("IMFCode.xlsx",sheet_name="Country")
indicators = pd.read_excel("IMFCode.xlsx",sheet_name="Indicators c-divided by service")
```

```python
dfog = dfog[dfog["FREQUENCY"] == "Annual"]
dfog = dfog[dfog["UNIT"] == "US dollar"]
dfog = dfog.iloc[:, 0:7]
dfog = dfog.drop(columns=["FREQUENCY","UNIT"])
dfog
```

Credit/Revenue Tables

```python
df = dfog[dfog["BOP_ACCOUNTING_ENTRY"] == "Credit/Revenue"]
df = df.drop(columns=["BOP_ACCOUNTING_ENTRY"])
df
```

```python
services = dfog[dfog['INDICATOR'] == "Services"].copy()
services = services[services["BOP_ACCOUNTING_ENTRY"] == "Credit/Revenue"]
services = services.drop(columns=["BOP_ACCOUNTING_ENTRY"])
services = services.rename(columns={'OBS_VALUE': 'SERVICE_VALUE'})


conn = sqlite3.connect(r"C:\IFs\RUNFILES\IFsDataImport - Copy (10).db")
cursor = conn.cursor()

for _, row in indicators.iterrows():
    indicator_id = row['Indicator']
    table_name = row['Variable']
    sql_table_name = f"Series{table_name}"

    ind_df = df[df['INDICATOR'] == indicator_id].copy()

    merged = ind_df.merge(
        services[['COUNTRY', 'TIME_PERIOD', 'SERVICE_VALUE']],
        on=['COUNTRY', 'TIME_PERIOD'],
        how='left'
    )

    merged['RESULT'] = (merged['OBS_VALUE'] / merged['SERVICE_VALUE'] * 100).round(3)
    merged = merged.dropna(subset=['TIME_PERIOD'])
    merged['TIME_PERIOD'] = merged['TIME_PERIOD'].astype(str).str.strip()
    
    final = merged[['COUNTRY', 'TIME_PERIOD', 'RESULT']]
    
    final = merged[['COUNTRY', 'TIME_PERIOD', 'RESULT']].pivot_table(
        index='COUNTRY',
        columns='TIME_PERIOD',
        values='RESULT'
    ).reset_index()   
     
    final = country.merge(final, on='COUNTRY', how='left')
    final = final.drop(columns=['COUNTRY'])
    final.columns = [str(c) for c in final.columns]

    year_columns = [col for col in final.columns if str(col).isdigit()]
    final['Earliest'] = final[year_columns].bfill(axis=1).iloc[:, 0]
    final['MostRecent'] = final[year_columns].ffill(axis=1).iloc[:, -1]
    final[year_columns + ['Earliest', 'MostRecent']] = final[year_columns + ['Earliest', 'MostRecent']].round(3)

    columns_sql = ['Country VARCHAR(255)', 'FIPS_CODE VARCHAR(255)']
    columns_sql += [f'"{col}" DOUBLE(53)' for col in year_columns + ['Earliest', 'MostRecent']]
    create_table_sql = f"CREATE TABLE '{sql_table_name}' ({', '.join(columns_sql)})"
    cursor.execute(f'DROP TABLE IF EXISTS "{sql_table_name}"')
    cursor.execute(create_table_sql)

    final.to_sql(sql_table_name, conn, if_exists='append', index=False)
    print(f"Imported: {sql_table_name}")

conn.commit()
conn.close()
```
