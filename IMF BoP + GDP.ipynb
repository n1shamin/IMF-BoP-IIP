```python
df = dfog[dfog["BOP_ACCOUNTING_ENTRY"] == "Credit/Revenue"]
df = df.drop(columns=["BOP_ACCOUNTING_ENTRY"])
df
```

```python
country = pd.read_excel("IMFCode.xlsx",sheet_name='Country')
```

```python
#sometimes one doesnt work
#df['OBS_VALUE'] = df['OBS_VALUE']*1000000
df.loc[df['SCALE'] == 'Millions', 'OBS_VALUE'] *= 1_000_000

df = df.dropna(subset=['TIME_PERIOD'])
df['TIME_PERIOD'] = df['TIME_PERIOD'].astype(str).str.strip()

workertransferrec = df[df['INDICATOR'] == "Secondary income, Financial corporations, nonfinancial corporations, households, and NPISHs"]
workertransferrec = workertransferrec.drop(columns="INDICATOR")
workertransferrec = workertransferrec.pivot(index='COUNTRY',columns='TIME_PERIOD',values='OBS_VALUE')
workertransferrec = workertransferrec.iloc[:,8:77]
workertransferrec = workertransferrec.reset_index()
workertransferrec.columns.name = None 
workertransferrec = country.merge(workertransferrec, on='COUNTRY', how='left')
workertransferrec = workertransferrec.drop(columns=['COUNTRY'])

workercomrec = df[df['INDICATOR'] == "Compensation of employees"]
workercomrec = workercomrec.drop(columns="INDICATOR")
workercomrec = workercomrec.pivot(index='COUNTRY',columns='TIME_PERIOD',values='OBS_VALUE')
workercomrec = workercomrec.iloc[:,8:73]
workercomrec = workercomrec.reset_index()
workercomrec.columns.name = None 
workercomrec = country.merge(workercomrec, on='COUNTRY', how='left')
workercomrec = workercomrec.drop(columns=['COUNTRY'])

incsecrec = df[df['INDICATOR'] == "Secondary income"]
incsecrec = incsecrec.drop(columns="INDICATOR")
incsecrec = incsecrec.pivot(index='COUNTRY',columns='TIME_PERIOD',values='OBS_VALUE')
incsecrec = incsecrec.iloc[:,12:77]
incsecrec = incsecrec.reset_index()
incsecrec.columns.name = None 
incsecrec = country.merge(incsecrec, on='COUNTRY', how='left')
incsecrec = incsecrec.drop(columns=['COUNTRY'])
```

```python
gdp = pd.read_excel("GDP2.xlsx",skiprows=4)
gdp = gdp.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
countrygdp = pd.read_excel("IMFCode.xlsx",sheet_name="WDICountry")
gdp = countrygdp.merge(gdp, on="Country Name", how="left")
gdp = gdp.drop(columns=["Country Name"])
gdp
```

```python
with pd.ExcelWriter('GDP+IMFCredit1.xlsx', engine='xlsxwriter') as writer:
    workertransferrec.to_excel(writer, sheet_name='workertransferrec')
    workbook = writer.book
    worksheet1 = workbook.add_worksheet('workertransferrec1')
    workercomrec.to_excel(writer, sheet_name='workercomrec')
    workbook = writer.book
    worksheet2 = workbook.add_worksheet('workercomrec1')
    incsecrec.to_excel(writer, sheet_name='incsecrec')
    workbook = writer.book
    worksheet3 = workbook.add_worksheet('incsecrec1')
    gdp.to_excel(writer, sheet_name='GDP')
```

```python
#credit
workertransferrec1 = pd.read_excel("GDP+IMFCredit1.xlsx",sheet_name="workertransferrec1")
year_columns = [col for col in workertransferrec1.columns if str(col).isdigit()]
workertransferrec1['Earliest'] = workertransferrec1[year_columns].bfill(axis=1).iloc[:, 0]
workertransferrec1['MostRecent'] = workertransferrec1[year_columns].ffill(axis=1).iloc[:, -1]
workertransferrec1[year_columns + ['Earliest', 'MostRecent']] = workertransferrec1[year_columns + ['Earliest', 'MostRecent']].round(4)

workercomrec1 = pd.read_excel("GDP+IMFCredit1.xlsx",sheet_name="workercomrec1")
year_columns = [col for col in workercomrec1.columns if str(col).isdigit()]
workercomrec1['Earliest'] = workercomrec1[year_columns].bfill(axis=1).iloc[:, 0]
workercomrec1['MostRecent'] = workercomrec1[year_columns].ffill(axis=1).iloc[:, -1]
workercomrec1[year_columns + ['Earliest', 'MostRecent']] = workercomrec1[year_columns + ['Earliest', 'MostRecent']].round(4)

incsecrec1 = pd.read_excel("GDP+IMFCredit1.xlsx",sheet_name="incsecrec1")
year_columns = [col for col in incsecrec1.columns if str(col).isdigit()]
incsecrec1['Earliest'] = incsecrec1[year_columns].bfill(axis=1).iloc[:, 0]
incsecrec1['MostRecent'] = incsecrec1[year_columns].ffill(axis=1).iloc[:, -1]
incsecrec1[year_columns + ['Earliest', 'MostRecent']] = incsecrec1[year_columns + ['Earliest', 'MostRecent']].round(4)

#debit
incpayment1 = pd.read_excel("GDP+IMFDebit.xlsx",sheet_name="incpayment1")
year_columns = [col for col in incpayment1.columns if str(col).isdigit()]
incpayment1['Earliest'] = incpayment1[year_columns].bfill(axis=1).iloc[:, 0]
incpayment1['MostRecent'] = incpayment1[year_columns].ffill(axis=1).iloc[:, -1]
incpayment1[year_columns + ['Earliest', 'MostRecent']] = incpayment1[year_columns + ['Earliest', 'MostRecent']].round(4)

workertransferpaid1 = pd.read_excel("GDP+IMFDebit.xlsx",sheet_name="workertransferpaid1")
year_columns = [col for col in workertransferpaid1.columns if str(col).isdigit()]
workertransferpaid1['Earliest'] = workertransferpaid1[year_columns].bfill(axis=1).iloc[:, 0]
workertransferpaid1['MostRecent'] = workertransferpaid1[year_columns].ffill(axis=1).iloc[:, -1]
workertransferpaid1[year_columns + ['Earliest', 'MostRecent']] = workertransferpaid1[year_columns + ['Earliest', 'MostRecent']].round(4)

workercompaid1 = pd.read_excel("GDP+IMFDebit.xlsx",sheet_name="workercompaid1")
year_columns = [col for col in workercompaid1.columns if str(col).isdigit()]
workercompaid1['Earliest'] = workercompaid1[year_columns].bfill(axis=1).iloc[:, 0]
workercompaid1['MostRecent'] = workercompaid1[year_columns].ffill(axis=1).iloc[:, -1]
workercompaid1[year_columns + ['Earliest', 'MostRecent']] = workercompaid1[year_columns + ['Earliest', 'MostRecent']].round(4)

#net
net1 = pd.read_excel("GDPIMFNet.xlsx",sheet_name="net1")
year_columns = [col for col in net1.columns if str(col).isdigit()]
net1['Earliest'] = net1[year_columns].bfill(axis=1).iloc[:, 0]
net1['MostRecent'] = net1[year_columns].ffill(axis=1).iloc[:, -1]
net1[year_columns + ['Earliest', 'MostRecent']] = net1[year_columns + ['Earliest', 'MostRecent']].round(4)
```

```python
#sql table; change table name each time you import 
conn = sqlite3.connect(r'C:\IFs\RUNFILES\IFsDataImport - Copy (9).db')
cursor = conn.cursor()
table_name = 'SeriesXWorkerTransferReceived%GDP'
columns_sql = ['Country VARCHAR(255)', 'FIPS_CODE VARCHAR(255)']
columns_sql += [f'"{col}" DOUBLE(53)' for col in year_columns + ['Earliest', 'MostRecent']]
create_table_sql = f"CREATE TABLE '{table_name}' ({', '.join(columns_sql)})"

workertransferrec1.to_sql(table_name, conn, if_exists='replace', index=False)

conn.commit()
conn.close()
```
